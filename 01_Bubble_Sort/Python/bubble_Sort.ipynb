```python
def bubble_sort(array):
    n = len(array)
    swapped = True
    swap_counter = 0
    swap_positions = []
    
    while swapped:
        swapped = False
        
        for i in range(1, n):
            if array[i - 1] > array[i]:
                
                array[i - 1], array[i] = array[i], array[i - 1]
                swapped = True
                
                swap_counter += 1
                
                swap_positions.append(list(array))
                
    return array, swap_counter, swap_positions            
```

### Exemplo de uso

Temos uma lista de entrada para ordenação:
> 64, 34, 25, 12, 22, 11, 90

```python

input_list = [64, 34, 25, 12, 22, 11, 90]
sorted_list, swaps, swap_positions = bubble_sort(input_list.copy())
```

Mostra a Ordenação

```python
print(f"Lista ordenada: {sorted_list}")
print(f"Número de Trocas: {swaps}")
```
